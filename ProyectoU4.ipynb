# ***Metodología***

Este proyecto se basa en un enfoque cuantitativo, empleando análisis estadísticos para evaluar la influencia de diferentes factores en el rendimiento académico de los estudiantes. La recolección de datos se realizó mediante un conjunto de variables relacionadas con hábitos personales, entorno familiar, características escolares y aspectos socioculturales, que fueron analizadas a través de pruebas estadísticas para identificar relaciones significativas y diferencias entre grupos. Entre los métodos utilizados, se comienza con el análisis de varianza ANOVA, que permitió comparar las medias del puntaje académico entre distintos niveles de factores categóricos como la participación parental, la calidad del profesorado y el acceso a recursos educativos. Este método ayuda a determinar si existen diferencias significativas entre grupos. Despues de esto, se aplicó la prueba de Tukey para realizar comparaciones múltiples entre grupos, identificando cuáles de ellos presentaban diferencias significativas en sus medias. Esta técnica es especialmente útil para controlar el error tipo I (ocurre cuando se rechaza una hipótesis nula que en realidad es verdadera). En conjunto, esta metodología permitió identificar patrones significativos y diferencias clave en el rendimiento académico de los estudiantes, proporcionando una base sólida para interpretar los factores que influyen en su éxito escolar.

# ***Variable: participacion_parental***

La variable "participación parental" mide el nivel de participación de los padres en la educación de sus hijos, categorizándola en tres niveles: bajo, medio y elevado. Esta variable se analiza en relación con el puntaje del examen para determinar cómo el apoyo parental impacta en el rendimiento académico de los estudiantes.

```python
df['participacion_parental'].unique()
```

```python
#Fátima Beneranda Marray(['bajo', 'medio', 'elevadgaña Meléndez

# Si todos los grupos tienen una distribución normal
import scipy.stats as stats

nivel_de_significancia = 0.05

# Realizar ANOVA
anova_result = stats.f_oneway(
    df[df['participacion_parental'] == 'bajo']['puntaje_examen'],
    df[df['participacion_parental'] == 'medio']['puntaje_examen'],
    df[df['participacion_parental'] == 'elevado'] ['puntaje_examen'],

)

# Interpretar los resultados
print("Resultados del ANOVA:")
print(f"F-Estadístico: {anova_result.statistic:.2f}")
print(f"Valor p: {anova_result.pvalue:.5f}\n")

if anova_result.pvalue < nivel_de_significancia:
    print("Conclusión: Hay diferencias significativas entre al menos dos tratamientos.")
else:
    print("Conclusión: No hay diferencias significativas entre los tratamientos.")

```

Al haber diferencia significativa en al menos 2 variables en al menos dos niveles de participación parental, se sugiere que el grado de apoyo de los padres influye de manera importante en el rendimiento académico de los estudiantes.

```python
#Fátima Beneranda Magaña Meléndez

# Si todos los grupos tienen una distribución normal.
# Si no cumplen con la normalidad, toman el resultado de la prueba de Kruskal
#     y pueden usar los gráficos de Tukey para tomar una decisión (ignorando la tabla).
#.    Estrictamente hablando, tendrían que realizar la prueba de Dunn.
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['puntaje_examen'], groups=df['participacion_parental'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Gráfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Grupos", xlabel="Medias")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Múltiples comparaciones entre todos los pares (Tukey)")

```

Se puede observar de manera más clara que existen diferencias significativas entre todos los pares de grupos de participación parental bajo, medio, y elevado. El grafico muestra que el grupo con baja participación parental tiene puntajes más bajos que los grupos con participación media y elevada. Además, la diferencia entre los grupos de participación elevada y media también es bastante notoria.
