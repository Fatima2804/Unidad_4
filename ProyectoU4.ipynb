# ***Variable: participacion_parental***

La variable "participacin parental" mide el nivel de participacin de los padres en la educacin de sus hijos, categorizndola en tres niveles: bajo, medio y elevado. Esta variable se analiza en relacin con el puntaje del examen para determinar cmo el apoyo parental impacta en el rendimiento acadmico de los estudiantes.

```python
df['participacion_parental'].unique()
```

```python
#Ftima Beneranda Marray(['bajo', 'medio', 'elevadgaa Melndez

# Si todos los grupos tienen una distribucin normal
import scipy.stats as stats

nivel_de_significancia = 0.05

# Realizar ANOVA
anova_result = stats.f_oneway(
    df[df['participacion_parental'] == 'bajo']['puntaje_examen'],
    df[df['participacion_parental'] == 'medio']['puntaje_examen'],
    df[df['participacion_parental'] == 'elevado'] ['puntaje_examen'],

)

# Interpretar los resultados
print("Resultados del ANOVA:")
print(f"F-Estadstico: {anova_result.statistic:.2f}")
print(f"Valor p: {anova_result.pvalue:.5f}\n")

if anova_result.pvalue < nivel_de_significancia:
    print("Conclusin: Hay diferencias significativas entre al menos dos tratamientos.")
else:
    print("Conclusin: No hay diferencias significativas entre los tratamientos.")

```

Al haber diferencia significativa en al menos 2 variables en al menos dos niveles de participacin parental, se sugiere que el grado de apoyo de los padres influye de manera importante en el rendimiento acadmico de los estudiantes.

```python
#Ftima Beneranda Magaa Melndez

# Si todos los grupos tienen una distribucin normal.
# Si no cumplen con la normalidad, toman el resultado de la prueba de Kruskal
#     y pueden usar los grficos de Tukey para tomar una decisin (ignorando la tabla).
#.    Estrictamente hablando, tendran que realizar la prueba de Dunn.
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['puntaje_examen'], groups=df['participacion_parental'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Grfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Grupos", xlabel="Medias")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Mltiples comparaciones entre todos los pares (Tukey)")

```

Se puede observar de manera ms clara que existen diferencias significativas entre todos los pares de grupos de participacin parental bajo, medio, y elevado. El grafico muestra que el grupo con baja participacin parental tiene puntajes ms bajos que los grupos con participacin media y elevada. Adems, la diferencia entre los grupos de participacin elevada y media tambin es bastante notoria.
